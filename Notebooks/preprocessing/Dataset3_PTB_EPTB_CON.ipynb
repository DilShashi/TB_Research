```python
# Import necessary library
import pandas as pd

# Set the file path for the input TXT file
file_path = r"C:\Users\dilan\Desktop\TB research\TB_Research\Datasets\raw\GSE83456_series_matrix.txt"

# Load the TXT file into a Pandas DataFrame, skipping metadata lines
dataset3 = pd.read_csv(file_path, sep='\t', skiprows=41)
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_dataset3 = dataset3.T

# Reset the index to make the transposed column names into a header
transpose_dataset3.columns = transpose_dataset3.iloc[0]

# Drop the first row (index 0) which has the headers duplicated
transpose_dataset3 = transpose_dataset3.drop(transpose_dataset3.index[0])

# Display the first few rows of the corrected transposed data
transpose_dataset3.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_dataset3.info()
```

## Handling Duplicate Rows

```python
transpose_dataset3.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

### Handling Duplicate Columns

```python
# Identify columns that start with '!Sample_characteristics_ch1' by their index positions
# The duplicate columns are in positions known by inspection with indices 8, 9, 10, 11 and 12
characteristics_indices = [8, 9, 10, 11, 12]

# Extract and rename each column using its index position
transpose_dataset3['!Sample_gender'] = transpose_dataset3.iloc[:, characteristics_indices[0]]
transpose_dataset3['!Sample_ethnicity'] = transpose_dataset3.iloc[:, characteristics_indices[1]]
transpose_dataset3['!Sample_age'] = transpose_dataset3.iloc[:, characteristics_indices[2]]
transpose_dataset3['!Sample_disease_state'] = transpose_dataset3.iloc[:, characteristics_indices[3]]
transpose_dataset3['!Sample_tissue'] = transpose_dataset3.iloc[:, characteristics_indices[4]]

# Drop the original duplicated columns by their indices
transpose_dataset3.drop(transpose_dataset3.columns[characteristics_indices], axis=1, inplace=True)

# Reorder the columns to place the new columns at the front
# Create a new order by specifying the newly created columns first
new_column_order = ['!Sample_gender', '!Sample_ethnicity', '!Sample_age', '!Sample_disease_state', '!Sample_tissue'] + \
                   [col for col in transpose_dataset3.columns if col not in ['!Sample_gender', '!Sample_ethnicity', '!Sample_age', '!Sample_disease_state', '!Sample_tissue']]

# Reassign columns to the DataFrame in the new order
transpose_dataset3 = transpose_dataset3[new_column_order]

```

```python
# Display the first few rows of the transposed data
transpose_dataset3.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_dataset3.columns[transpose_dataset3.nunique() == 1]

# Drop those columns
data3 = transpose_dataset3.drop(columns=columns_to_drop)
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_disease_state" column
unique_categories = data3["!Sample_disease_state"].unique()

# Display the unique categories
unique_categories
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower()
    if 'disease state: ptb' in source_name:
        return 'Pulmonary TB'
    elif 'disease state: eptb' in source_name:
        return 'Extra Pulmonary TB'
    elif 'disease state: hc' in source_name:
        return 'Healthy Control'
    else:
        return 'Other Disease'  # Other lung disease like 'Lung Cancer'

# Apply the function to the "!Sample_disease_state" column
data3['TB_Status'] = data3['!Sample_disease_state'].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(data3['TB_Status'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = data3.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    data3[col] = cap_outliers(data3[col])

# Display the random 5 rows to verify the changes
data3.sample(5)
```
