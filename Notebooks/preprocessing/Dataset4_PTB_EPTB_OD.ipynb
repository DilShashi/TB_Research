```python
# Import necessary library
import pandas as pd

# Set the file path for the input TXT file
file_path = r"C:\Users\dilan\Desktop\TB research\GSE144127_series_matrix.txt"

# Load the TXT file into a Pandas DataFrame, skipping metadata lines
dataset4 = pd.read_csv(file_path, sep='\t', skiprows=37)
```

## Get the transpose of the dataset

```python
# Transpose the dataset
transpose_dataset4 = dataset4.T

# Reset the index to make the transposed column names into a header
transpose_dataset4.columns = transpose_dataset4.iloc[0]

# Drop the first row (index 0) which has the headers duplicated
transpose_dataset4 = transpose_dataset4.drop(transpose_dataset4.index[0])

# Display the first few rows of the corrected transposed data
transpose_dataset4.head()
```

```python
# Check basic information about the dataset, including data types and non-null counts
transpose_dataset4.info()
```

## Handling Duplicate Rows

```python
transpose_dataset4.drop_duplicates(subset=['ID_REF'], keep='first', inplace=True)
```

### Handling Duplicate Columns

```python
# Identify columns that start with '!Sample_characteristics_ch1' by their index positions
# The duplicate columns are in positions known by inspection with indices 8, 9, 10, 11 and 12
characteristics_indices = [12]

# Extract and rename each column using its index position
transpose_dataset4['!Sample_type'] = transpose_dataset4.iloc[:, characteristics_indices[0]]

# Drop the original duplicated columns by their indices
transpose_dataset4.drop(transpose_dataset4.columns[characteristics_indices], axis=1, inplace=True)

# Reorder the columns to place the new columns at the front
# Create a new order by specifying the newly created columns first
new_column_order = ['!Sample_type'] + \
                   [col for col in transpose_dataset4.columns if col not in ['!Sample_type']]

# Reassign columns to the DataFrame in the new order
transpose_dataset4 = transpose_dataset4[new_column_order]

```

```python
# Display the first few rows of the transposed data
transpose_dataset4.head()
```

## Unnecessary Feature Elemination

```python
# Identify columns where all values are the same
columns_to_drop = transpose_dataset4.columns[transpose_dataset4.nunique() == 1]

# Drop those columns
data4 = transpose_dataset4.drop(columns=columns_to_drop)
```

```python
# Drop columns with more than 50% missing values
# Calculate threshold for dropping columns
threshold = len(data4) * 0.5
data4 = data4.dropna(thresh=threshold, axis=1)
```

```python
# Select the first 5 columns
first_5_columns = data4.columns[:5]

# Create an empty list to store the rows
rows = []

# Loop through each of the first 5 columns
for col in first_5_columns:
    # Get the data type of the column using the Series dtype
    data_type = data4[col].dtypes  # Access the dtype via .dtypes (not .dtype)

    # Get the first two row values of the column
    row_value_1 = data4[col].iloc[0]
    row_value_2 = data4[col].iloc[1]

    # Append the information as a row to the list
    rows.append({
        'Feature': col,
        'Data Type': data_type,
        'Row Value 1': row_value_1,
        'Row Value 2': row_value_2
    })

# Convert the list of rows into a DataFrame
result_df = pd.DataFrame(rows)

# Display the result
result_df
```

## Creating the Target Column - TB_Status

```python
# Get unique categories in the "!Sample_type" column
unique_categories = data4["!Sample_type"].unique()

# Display the unique categories
unique_categories
```

```python
data4.sample(5)
```

```python
# Define a function to categorize each sample type
def categorize_sample_type(source_name):
    source_name = source_name.lower().strip()
    if 'eptborptb: ptb' in source_name or 'site of infection: lung' in source_name:
        return 'Pulmonary TB'
    elif 'eptborptb: eptb' in source_name or 'site of infection: spine' in source_name:
        return 'Extra Pulmonary TB'
    elif 'eptborptb: ptb_eptb' in source_name:
        return 'PTB and EPTB'
    elif 'smear microscopy: true' in source_name:
        return 'TB'
    else:
        return 'Other Disease'  # Other lung disease like 'Lung Cancer'

# Apply the function to the "!Sample_disease_state" column
data4['TB_Status'] = data4['!Sample_type'].apply(categorize_sample_type)

# Display unique values in the new 'TB_Status' column to verify
print(data4['TB_Status'].unique())
```

## Handling Outliers

```python
import numpy as np

# Select only the numeric columns
numeric_columns = data4.select_dtypes(include=['float64', 'int64']).columns

# Define a function to cap outliers
def cap_outliers(series):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values outside the bounds to the respective bounds
    return np.where(series < lower_bound, lower_bound,
                    np.where(series > upper_bound, upper_bound, series))

# Apply the function to each numeric column
for col in numeric_columns:
    data4[col] = cap_outliers(data4[col])

# Display the random 5 rows to verify the changes
data4.sample(5)
```
